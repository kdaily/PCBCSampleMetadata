Function to add minimal annotations to bootstrap the rest.

```python
def update(x, syn):
    x.fileType = 'count'
    x.dataType = 'mRNA'
    x.sampleName = x.name.replace(".txt", "")
    x.UID = x.sampleName
    return syn.store(x, forceVersion=False)
```

```python
objs = objs.map(lambda x: update(x, syn))
```
