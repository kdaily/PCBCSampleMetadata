Proof of concept how output to ISA-Tab could work from annotations on Synapse files.

```python
import pandas as pd

import PandaTables
import PCBCTables
import synapseHelpers

import synapseclient

syn = synapseclient.Synapse(skip_checks=True)

syn.login(silent=True)
```

Get the BAM files, only those with UID set are ones that have the necessary annotations to continue.

```python
pdbam = PCBCTables.MRNABAMAnnotationQuery(syn=syn)
pdbam["ISAfileType"] = "Aligned"
pdbam = pdbam[pdbam['UID'].notnull()]
pdbam_idxd = pdbam#.set_index("UID")
```

Get the fastq files. They are currently missing UIDs because they did not have any annotations at all to add them in the bootstrapping procedure.

```python
pdfq = PCBCTables.MRNAFastqAnnotationQuery(syn=syn)
pdfq["ISAfileType"] = "Raw"
pdfq['UID'] = pdfq['name'].map(lambda x: x.replace('.fastq', ''))
pdfq_idxd = pdfq#.set_index("UID")
```

```python
pdfq.head()
```

Get the normalized files. These should be moved to another directory to simplify the query.

```python
normfiles = PCBCTables.NormalizedMatrixFileQuery(syn=syn)
normfiles = normfiles[normfiles['dataType'].notnull()]
```

Merge the files together.

```python
mg = pd.merge(pdfq_idxd, pdbam_idxd, left_on="UID", right_on="UID")
mg = pd.merge(mg, normfiles, left_on="dataType_x", right_on="dataType", how="left")
```

Do some table formatting to make it look more ISA-Tab like.

```python
colname_map = {"C4_Cell_Line_ID_y": "C4_Cell_Line_ID", "id_x": "fastq id", "name_x": "fastq filename",
               "ISAfileType_x": "fastq filetype", "ISAfileType_y": "bam filetype",  
               "id_y": "bam id", "name_y": "bam filename", "id": "normalized id", "name":
               "normalized filename", "Diffname_short_y": "Characteristics"}

keepcols = colname_map.keys() + ["UID"]
mg = mg[keepcols]

mg.rename(columns=colname_map, inplace=True)

# Reorder columns
mg = mg.reindex_axis(["C4_Cell_Line_ID", "Characteristics", "UID",
                      "fastq filetype", "fastq filename", "fastq id", 
                      "bam filetype", "bam filename", "bam id",
                      "normalized filename", "normalized id"],
                axis=1)
```
